```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
with open("/home/maria/MouseViT2/src/ci_accuracy_comparison.pkl", "rb") as f:
    results = pickle.load(f)

accuracies_real = np.array(results['real'])
accuracies_perm = np.array(results['perm'])

# Remove NaNs
mask = ~np.isnan(accuracies_real) & ~np.isnan(accuracies_perm)
accuracies_real = accuracies_real[mask]
accuracies_perm = accuracies_perm[mask]

# Plot
plt.figure(figsize=(10, 6))
sns.histplot(accuracies_real, label="Real Embeddings", color="blue", kde=True, stat="density", bins=20)
sns.histplot(accuracies_perm, label="Permuted Embeddings", color="red", kde=True, stat="density", bins=20, alpha=0.6)

plt.title("Confidence Interval Accuracy per Neuron")
plt.xlabel("Average Accuracy (prediction within CI)")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.savefig("ci_accuracy_histogram_comparison.png", dpi=300)
plt.show()
```

```python
print(f"Mean accuracy (real): {np.nanmean(accuracies_real):.4f}")
print(f"Mean accuracy (perm): {np.nanmean(accuracies_perm):.4f}")
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the saved log-likelihoods
with open("/home/maria/MouseViT2/src/event_only_loglik_cv.pkl", "rb") as f:
    data = pickle.load(f)

real = np.array(data["real"])
perm = np.array(data["perm"])
delta = real - perm

# Remove NaNs or infs (if any)
mask = np.isfinite(real) & np.isfinite(perm)
real = real[mask]
perm = perm[mask]
delta = delta[mask]

# Plot 1: Real vs Perm log-likelihood histograms
plt.figure(figsize=(9, 5))
sns.histplot(real, label='Real Embeddings', color='blue', stat='density', kde=True, bins=20, alpha=0.6)
sns.histplot(perm, label='Permuted Embeddings', color='red', stat='density', kde=True, bins=20, alpha=0.6)
plt.axvline(np.mean(real), linestyle='--', color='blue', label='Real Mean')
plt.axvline(np.mean(perm), linestyle='--', color='red', label='Perm Mean')
plt.title("Log-likelihood Distribution at Spikes (C = 10)")
plt.xlabel("Log-likelihood at y = 1")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.savefig("loglik_event_only_histogram.png", dpi=300)
plt.show()

# Plot 2: Delta histogram (real - perm)
plt.figure(figsize=(7, 5))
sns.histplot(delta, color='purple', kde=True, bins=20)
plt.axvline(np.mean(delta), linestyle='--', color='black', label=f"Mean  = {np.mean(delta):.3f}")
plt.title("Delta Log-likelihood (Real - Permuted)")
plt.xlabel(" Log-likelihood at Spikes (y = 1)")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.savefig("loglik_event_only_delta_histogram.png", dpi=300)
plt.show()
```
